Creates a function which calculates the correlation between trail stores and other stores

```python
def CalculateCorrelation(inputtable,trail_store,metric):
    output = pd.DataFrame({'Store 1':[],'Store 2':[],'Correlation':[]})
    a = inputtable[inputtable['STORE_NBR']==trail_store][metric]
    a.reset_index(drop = True, inplace = True)
    store_numbers = inputtable['STORE_NBR'].unique()
    for i in store_numbers:
        b = inputtable[inputtable['STORE_NBR']==i][metric]
        b.reset_index(drop = True, inplace = True)
        corr = b.corr(a)
        output = output.append({'Store 1':trail_store,'Store 2':i,'Correlation':corr},ignore_index=True)
    return output
```

```python
def CalculateMagnitude(inputtable,trail_store,metric):
    output = pd.DataFrame({'Store 1':[],'Store 2':[],'Magnitude':[]})
    a = inputtable[inputtable['STORE_NBR']==trail_store][metric]
    a.reset_index(drop = True, inplace = True)
    store_numbers = inputtable['STORE_NBR'].unique()
    for i in store_numbers:
        b = inputtable[inputtable['STORE_NBR']==i][metric]
        b.reset_index(drop = True, inplace = True)
        c = abs(a-b)
        # the smaller difference in magnitude, closer to zero
        d = np.mean(1-(c-min(c))/(max(c)-min(c)))
        output = output.append({'Store 1':trail_store,'Store 2':i,'Magnitude':d},ignore_index=True)
    return output
```

### Assessment of trial for trial strore 77

The trial period goes from the start of March 2019 to June 2019. We now want to see if there has been an uplift in overall chip sales.

```python
control = 233
trail = 77
metric = ['Total Sales']
assessment = summary[(summary['STORE_NBR']==control)|(summary['STORE_NBR']==trail)]
assessment_pre = assessment[assessment['month_year'] < '2019-02']
```

```python
trail_sum = assessment_pre[assessment['STORE_NBR']==trail][metric].sum()
control_sum = assessment_pre[assessment['STORE_NBR']==control][metric].sum()
scalingfactor = float(trail_sum/control_sum)
```

```python
scaled_control = assessment[assessment['STORE_NBR']==control]
scaled_control['Control Sales']=scaled_control['Total Sales']*scalingfactor
percentage_diff = scaled_control[['month_year','Control Sales']]
percentage_diff.reset_index(drop=True,inplace=True)
```

```python
trail_metric = assessment[assessment['STORE_NBR']==trail][metric]
trail_metric.reset_index(drop=True,inplace=True)
percentage_diff = pd.concat([percentage_diff,trail_metric],axis=1)
percentage_diff.rename(columns = {'Total Sales': 'Trail Sales'}, inplace = True)
percentage_diff.info()
```

```python
percentage_diff['Diff'] = abs(percentage_diff['Trail Sales']-percentage_diff['Control Sales'])/percentage_diff['Control Sales']
percentage_diff
```

Our null hypothesis is the trail period is the same as the pre-trail period. Let's calculate the standard deviation from the pre-trail period.

```python
from statistics import stdev
std_dev = stdev(percentage_diff[percentage_diff['month_year']<' 2019-02']['Diff'])
std_dev
```

Since we have 7 pre-trail month, degree of freedom = 7 - 1 = 6. The null hypothesis is the difference is 0 from Feb 2019 to Apr 2019.

```python
dof = 6
```

```python
percentage_diff['t_value']=(percentage_diff['Diff']-0)/std_dev
percentage_diff[(percentage_diff['month_year']>'2019-01')&(percentage_diff['month_year']<'2019-05')]
```

```python
# find the 95th percentile critival value
from scipy.stats import t

t.isf(0.05, dof)
```

We can see that the t-value in March 2019 and April 2019 is greater than the critical value, meaning we are able to reject the null hypothesis that the difference is 0 in March 2019 and April 2019.
